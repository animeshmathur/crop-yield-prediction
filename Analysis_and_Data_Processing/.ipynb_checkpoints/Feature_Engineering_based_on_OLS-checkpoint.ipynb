## Convert Production values to Log-Normal Form

During EDA we saw that log of 99.5% quantile of Production values had formed a perfect normal distribution.

```python
def convert_to_log(val):
    return np.log(np.abs(val) + 1)

def convert_from_log(val):
    if np.exp(val) == 1:
        return 0
    return np.exp(val) - 1
```

```python
np.log(-1)
```

```python
df['log_Production'] = df['Production'].apply(convert_to_log)
```

```python
## Taking data with only 99.5% quantile of log_Production
df = df.loc[df['log_Production'] <= df['log_Production'].quantile(0.995)]
```

```python
with open('./latlong_custerer_6.pkl', 'rb') as f:
    lat_long_clusterer = pickle.load(f)
```

```python
def identify_geo_region(data, clusterer):
    '''
    Returns dataframe with column 'Geo_Region' identified from values of 'Lat' and 'Long' for each row.
    
    Parameters:
    data - Data frame which should have 'Lat' and 'Long' columns.
    clusterer - Clustering model for Lat and Long.
    '''
    data['Geo_Region'] = clusterer.predict(data[['Lat', 'Long']])
    
    region_map = {}
    
    for i, region in enumerate(df['Geo_Region'].unique()):
        region_map[i] = f'R{i}'
        
    data['Geo_Region'] = data['Geo_Region'].map(region_map).copy()
    
    return data
```

```python
def ols_summary(cols, target, data):
    formula_str = f"{target} ~ "
    for col in cols:
        formula_str += f" + {col}"
    return smf.ols(formula=formula_str, data=data).fit().summary()
```

Going forward, we will take previous formula as base to check for improvement in Adj. R-squared

Lets see if Adj R-squared value increses on log transformation of any other feature

```python
# This function find those columns which show improvement in Adj. R-squared value upon log transformation
def get_log_cols(cols, target, data):
    max_rsquared_adj = 0
    log_cols = []
    
    # Adj. R-squared without log transformation
    formula_str = f"{target} ~ "
    for col in cols:
        formula_str += f" + {col}"
    max_rsquared_adj = smf.ols(formula=formula_str, data=data).fit().rsquared_adj
    
    improvement_log = [max_rsquared_adj]
    
    # Check improvement in Adj. R-squared with log transformation on each column
    for log_col in cols:
        formula_str = f"{target} ~ "
        for col in cols:
            if col == log_col or (col in log_cols):
                formula_str += f" + np.log({col} + 1)"
            else:
                formula_str += f" + {col}"
        rsquared_adj = smf.ols(formula=formula_str, data=data).fit().rsquared_adj
        if max_rsquared_adj < rsquared_adj:
            log_cols.append(log_col)
            max_rsquared_adj = rsquared_adj
            improvement_log.append(rsquared_adj)
    return log_cols, improvement_log
```

```python
get_log_cols(['GW_depth_Min', 'GW_depth_Max', 'Wells_depth_0_to_2',
           'Wells_depth_2_to_5', 'Wells_depth_5_to_10', 'Wells_depth_10_to_20',
           'Wells_depth_20_to_40', 'Wells_depth_40_and_above', 'Precipitation',
           'Solar_Radiation', 'Surface_Pressure', 'Humidity', 'Temp_Max',
           'Temp_Min', 'Wind_Speed', 'Area'], 'log_Production', df)
```

From above analysis, we see that only the logarithmic transformation of Area column gives significantly high Adj. R-squared.
