```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

import pickle
```

```python
df = raw_data.copy()
```

```python
# Drop % columns as they are collinear
df.drop(['in %', 'in %.1', 'in %.2', 'in %.3', 'in %.4', 'in %.5'], axis=1, inplace=True)
```

```python
df.describe(include='all')
```

```python
df['No of wells monitored'].unique()
```

```python
# extract count from the typo string
df['No of wells monitored'].loc[df['No of wells monitored'] == 'R3'] = 3
df['No of wells monitored'].loc[df['No of wells monitored'] == 'AR7'] = 7
df['No of wells monitored'].loc[df['No of wells monitored'] == 'GAR 5'] = 5
```

```python
df['No of wells monitored'].unique()
```

```python
# change dtype to integer for no.of wells
df['No of wells monitored'] = df['No of wells monitored'].astype(int)
```

```python
df.describe()
```

```python
# Convert -ve values to +ve
df['Minimum'] = np.abs(df['Minimum'])
df.describe()
```

```python
df[df['Maximum'] <= 0]
```

```python
# Swap Minumum and Maximum values for row at index 18
temp = df['Maximum'][18]
df['Maximum'][18] = df['Minimum'][18]
df['Minimum'][18] = np.abs(temp)
```

```python
df.describe()
```

```python
df_geo = df_cleaned.copy()
```

```python
import time
import urllib, json

geo_data = {}
requested_index = 0
```

```python
# read
geo_file = open('./data/geo_data', 'rb')
geo_data = pickle.load(geo_file)                      
geo_file.close()

if len(geo_data) == 0:
    for state, district in zip(df_cleaned['State'].iloc[requested_index:], df_cleaned['District'].iloc[requested_index:]):
        q = urllib.parse.quote(f"{district},{state}")
        req_url = f"http://geocode.xyz/{q}?json=1"
        with urllib.request.urlopen(req_url) as url:
            data = json.loads(url.read().decode())
            geo_data[district] = {"long": data['longt'], "lat": data['latt']}
            print(state, district, geo_data[district])
            requested_index += 1
        time.sleep(5)

    # write
    out_file = open('geo_data', 'ab')
    pickle.dump(geo_data, out_file)                      
    out_file.close() 
```

```python
df_geo['Lat'] = df_geo['District'].apply(lambda d: geo_data[d]['lat']).astype(np.float32)
df_geo['Long'] = df_geo['District'].apply(lambda d: geo_data[d]['long']).astype(np.float32)
```
