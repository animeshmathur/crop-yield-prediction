```python
df_apy.columns = ['State', 'District', 'Crop_Year', 'Season', 'Crop', 'Area', 'Production']
```

```python
# Join APY and GW dataframes based upon District
df_merged = df_gw.set_index(['State', 'District']).join(df_apy.set_index(['State', 'District']), how='inner')
```

```python
df_merged = df_merged.reset_index()
df_merged
```

```python
len(df_merged.District.unique())
```

```python
def duplicate_district_check(d, df):
    return df['State'][df['District'] == d].unique()

for d in df_merged['District'].unique():
    s = duplicate_district_check(d, df_merged)
    if len(s)>1:
        print(d, s)
```
